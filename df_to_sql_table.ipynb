```python
import pandas as pd
import sqlite3 as sql # db.close()
```

```python
db = sql.connect("newdb.db")
dbc = db.cursor()
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/refs/heads/main/Electricity_Production_By_Source.csv")
```

```python
df.head()
```

```python
df.to_sql("elect", db)
```

```python
cols = ['country', 'code', 'year', 'coal', 'gas', 'hydro', 'renewables', 'solar', 'oil', 'wind', 'nuclear']
df.columns = cols
```

```python
df.head()
```

```python
df.to_sql("electro", db)
```

```python
df.coal.head()
```

```python
dbc.execute("SELECT name FROM sqlite_master WHERE type = 'table';") # jadvallar haqida ma'lumot olish
dbc.fetchall() # query dan kelgan javobni olish
```

```python
dbc.execute("PRAGMA table_info(electro)")
dbc.fetchall()
```
